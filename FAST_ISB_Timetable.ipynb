# 3. Return all the classes of the particular subject (Ask from User) along with Time, Day and Room Number.

```python
subject = str(input("Enter the subject that you want to see the time table of: "))
```

```python
for i in time_table:
    mm = mon.where(mon==subject).dropna(axis = 0, how = 'all').dropna(axis = 1)
    tt = tue.where(tue==subject).dropna(axis = 0, how = 'all').dropna(axis = 1)
    ww = wed.where(wed==subject).dropna(axis = 0, how = 'all').dropna(axis = 1)
    tH = thu.where(thu==subject).dropna(axis = 0, how = 'all').dropna(axis = 1)
    ff = fri.where(fri==subject).dropna(axis = 0, how = 'all').dropna(axis = 1)

if not mm.empty:
    print("MONDAY \n", mm)
if not tt.empty:
    print("TUESDAY \n", tt)
if not ww.empty:
    print("WEDNESDAY \n", ww)
if not tH.empty:
    print("THURSDAY \n", tH)
if not ff.empty:
    print("FRIDAY \n", ff)
```

# 6. Which Classroom is less busy in overall week.

```python
mon_1 = mon.iloc[:35,:]
#mon_1 = mon.iloc[:38,:]                                                          #if we want to include classes held in rooms without name
sum_mon = mon_1[mon_1.isnull().sum(axis=1) == max(mon_1.isnull().sum(axis=1))]    #return the rows that have the maximum amount of missing values
sum_mon
```

```python
tue_1 = tue.iloc[:35,:]
#tue_1 = tue.iloc[:38,:]  if we want to include classes held in rooms without name
sum_tue = tue_1[tue_1.isnull().sum(axis=1) == max(tue_1.isnull().sum(axis=1))]
sum_tue
```

```python
wed_1 = wed.iloc[:35,:]
#wed_1 = wed.iloc[:38,:]  if we want to include classes held in rooms without name
sum_wed = wed_1[wed_1.isnull().sum(axis=1) == max(wed_1.isnull().sum(axis=1))]
sum_wed
```

```python
thu_1 = thu.iloc[:37,:]
sum_thu = thu_1[thu_1.isnull().sum(axis=1) == max(thu_1.isnull().sum(axis=1))]
sum_thu
```

```python
fri_1 = fri.iloc[:38,:]
sum_fri = fri_1[fri_1.isnull().sum(axis=1) == max(fri_1.isnull().sum(axis=1))]
sum_fri
```

```python
m = pd.concat([sum_mon,sum_tue,sum_wed,sum_thu, sum_fri], axis=1, ignore_index=True) 
m = m.dropna(how='all', axis=1)       #dropping columns where all the columns are = Nan
m
```

```python
sum_total = m[m.isnull().sum(axis=1) == max(m.isnull().sum(axis=1))]
print("Least busy classes are:", sum_total.index.values)
```
